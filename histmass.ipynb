```
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```
df = pd.read_csv('https://raw.githubusercontent.com/fisicadeparticulas/20190218-invariantmass-jordetm5/master/Dimuon_DoubleMu.csv', index_col=0)
df.head()
```

```
m=df.M
m=np.array(m)
plt.hist(m, 70, histtype="stepfilled", alpha=1)
plt.show()
```
